```python
#1.2. На складе лежат разные фрукты в разном количестве. Написать функцию, которая на вход принимает любое 
#количество названий фруктов и их количество, а возвращает общее количество фруктов на складе
#Пример: <finction_name>(apple=2, banana=12, pear=9, peach=5) -> 28
def fruits(**kwargs):
    total_fruits = sum(kwargs.values())
    return total_fruits

print(fruits(apples=2, pears=12, bananas=9, oranges=5))
```
